Importing the dataset and libraries.

Separating the year from the month values and assign each one to his own column.

```python
rain = rain.rename(columns={'Mês/Ano':'month_year'})
```

```python
rain.head()
```

```python
rain['year']= rain['month_year'].copy()
```

```python
rain.head()
```

```python
aux=[]
for i in range (0,len(rain['month_year'])):
    if rain['month_year'][i] !=0:
        aux = rain['month_year'][i]
        rain['month_year'][i] = aux[:3]

rain['month_year'].value_counts()
```

```python
aux=[]
for i in range (0,len(rain['year'])):
    if rain['year'][i] !=0:
        aux = rain['year'][i]
        rain['year'][i] = aux[-4:]
        rain['year'][i] = int(rain['year'][i])

rain['year'][:10]
```

Transforming the months into numerical values.

```python
mapping = {'jan': 1, 'fev': 2,'mar': 3,'abr': 4,'mai': 5,'jun':6, 'jul':7,'ago':8,'set':9,'out':10,'nov':11,'dez':12}

rain = rain.applymap(lambda s: mapping.get(s) if s in mapping else s)
rain.head()
```

```python
rain['month_year'].value_counts()
```

```python
rain = rain.rename(columns={'month_year':'month'})
```

This dataset is the rain registered at that station on that day so in order to merge with the other dataset, there is a need to reorder how the information is shown. The day needs to be in a single column matched with it's respective value. While doing so we will clear the null values.

```python
column_names = list(rain.columns.values)
column_names = column_names[3:-2]
```

```python
temporary_row= [[0,0,0,0,0,0]]
rain_transformed = pd.DataFrame(temporary_row,columns=['Código','Posto','month','year','rain','day'])
temporary=[]
for i in column_names:
    temporary = rain[['Código','Posto','month','year',i]]
    temporary = temporary.rename(columns={i:'rain'})
    temporary = temporary[temporary.rain != '-']
    day = [i for n in range(0,len(temporary['rain']))]
    temporary['day']= day
    rain_transformed = pd.concat([rain_transformed,temporary], axis=0, join='outer', ignore_index=False)

rain_transformed['day'] = rain_transformed.day.astype(int)
rain_transformed = rain_transformed.drop([0])
rain_transformed.head()
```
